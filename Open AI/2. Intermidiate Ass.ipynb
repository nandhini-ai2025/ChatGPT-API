```python
print("Instant Chatbot Ready! (type 'quit' to stop)\n")

while True:
    user_input = input("You: ").lower()

    if user_input == "quit":
        print("Bot: Goodbye ")
        break

    elif "hello" in user_input:
        print("Bot: Hello! How can I help you?")
    
    elif "ai" in user_input:
        print("Bot: AI means Artificial Intelligence, machines thinking like humans.")
    
    elif "machine learning" in user_input:
        print("Bot: Machine Learning is a subset of AI that learns from data.")
    
    elif "interview" in user_input:
        print("Bot: Prepare basics, projects, and practice communication.")
    
    elif "python" in user_input:
        print("Bot: Python is a popular programming language for AI and data science.")
    
    else:
        print("Bot: Sorry, I don't understand. Try asking something else.")
```
