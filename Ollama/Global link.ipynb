```python
import ollama
import gradio as gr

MODEL = "llama3"

# 🔥 Warm-up
ollama.chat(model=MODEL, messages=[{"role": "user", "content": "hi"}])

# ✅ FIXED FUNCTION (2 parameters)
def chat(user_input, history):
    response = ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": user_input}],
        options={
            "num_predict": 100,   # fast output
            "temperature": 0.5
        }
    )

    return response["message"]["content"]

# Gradio UI
gr.ChatInterface(fn=chat, title="⚡ Fast Chatbot").launch()
```
